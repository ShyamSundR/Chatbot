```python
import os
import chainlit as cl
from langchain import HuggingFaceHub, PromptTemplate, LLMChain
```

```python
from getpass import getpass
HUGGINGFACEHUB_API_TOKEN = getpass()
os.environ['HUGGINGFACEHUB_API_TOKEN'] = HUGGINGFACEHUB_API_TOKEN
```

```python
model_id = "gpt2-medium"
conv_model = HuggingFaceHub(huggingfacehub_api_token=os.environ['HUGGINGFACEHUB_API_TOKEN'], repo_id=model_id, model_kwargs={"temperature":0.8, "max_new_tokens":200})

```

```python
template = """Your are a helpful AI assistant that generates emails for the user 

{query}
"""

prompt = PromptTemplate(template=template, input_variables=['query'])
```

```python
conv_chain = LLMChain(llm=conv_model,
                     prompt=prompt,
                     verbose=True)
```

We can modify the print statement to check different outputs


```python
print(conv_chain.run("write a mail for sick leave "))
```
